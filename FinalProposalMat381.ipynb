# Dataset
Airbnb data : [Dataset](https://www.kaggle.com/datasets/alessiocrisafulli/airbnb-italy?resource=download). Airbnb is an application that helps customers who want to stay outside their own home to stay elsewhere for a fee.This dataset we got from Kaggle includes information about accommodation areas in some cities in Italy. Our dataset is information about the accommodation (location, number of bathrooms, number of toilets, number of rooms, daily fee), feedback of the user about the house (cleanliness score, accuracy score of the house, score of the location of the house, check-in score of the house) and contains information about the host (how many years he has been renting his house, response time to messages, having a verified account, host response rate).

Host_id : Unique id given to each host <br>
Host_since : When the landlord starts renting their first home <br>
Cleanliness_score : User-rated score of house cleanliness <br>
Accuracy score of the house : User score of the accuracy of the house's said features <br>
Score of the location of the house : User score of house location <br>
Check-in score of the house : Ease of check-in, score of time by user <br>
Response time to messages : Response time to the host's messages <br>
Having a verified account : The value of whether the host is an airbnb-confirmed account(1 or 0) <br>
Host response rate : Response rate of the host to messages sent by users <br>
Location,number of bathrooms, number of toilets, number of rooms, daily fee <br>

# Description of the Problem
By using this dataset, we will analyze how airbnb newcomers can more easily rent their homes in an effective and convenient way.<br>
What matters more to people who rent houses?<br>
Is it more important that the house has an escape room or how many beds it has?<br>
Which features are preferred more?<br>
What effect does the location of the house have on the holding of the house?<br>
In which location are the houses rented the most?<br>
What is the effect of the pricing policy on the rental of the house?<br>
To what extent do people take into househoulder response time to messages?<br>
With the answers we obtained by asking these questions to our dataset, we aim to identify what is more important for customers and share these findings with the households, and offer them the opportunity to rent their homes more easily.

# Project Planning
## Project Pieces
Since our dataset includes data that we are not interested in, such as user comments, we will first start by extracting the data that will be useful to us. Afterwards, we will search for answers to these questions by using our relevant data according to the questions we have determined.While searching for answers to these questions, we will determine the most appropriate ML algorithm and continue on it.The ML algorithms we will use are Logistic Regression, Linear Regression and Decision Tree algorithms.We will use Logistic Regression and Linear Regression to find the most accurate model by revealing which data are related to each other.We will use Decision Tree to correctly classify our dataset.Furthermore,we aim to visualize our data by using the Folium library.Finally, our target is to share these answers with the households, giving them the opportunity to rent a house more easily, while at the same time aiming to offer a more comfortable accommodation to the customers.<br>
